```python
import pandas as pd
df_cc = pd.read_excel("Covered Calls.xlsx")
print(df_cc.shape)
df_cc.head(2)
```

```python
df_temp = df_cc.loc[df_cc['Expiration Date']=="2024-02-16"]
closes = []
for i in df_temp.index:
    symbol = df_temp.loc[i,'Ticker']
    response = tiny.market_data.get_history(symbol=symbol,start="2024-02-16",end="2024-02-16")
    close = 0
    try:
        r = response.json()
        print(symbol,r['history']['day']['date'],r['history']['day']['close'])
        close = r['history']['day']['close']
    except Exception as e:
        pass
    finally:
        closes.append(close)
df_temp['Close Price'] = closes
df_temp.to_excel("Covered Calls Close 2024-02-16.xlsx",index=False)
df_temp.head()
```

```python
response = tiny.market_data.get_company_dividends(symbols="MSFT,MO,KMI")
response
```

```python
import pandas as pd
df_div = pd.DataFrame()
response.json()
for r in response.json():
    for result in r['results']:
        df_temp = pd.DataFrame(result['tables']['cash_dividends'])
        df_temp['symbol'] = r['request']
        df_temp['type'] = result['type']
        df_div = pd.concat([df_div,df_temp],ignore_index=True)
df_div.shape
```
