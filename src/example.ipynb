```python
response = tiny.market_data.get_quotes(symbols="AAPL,MSFT,NVDA,SPY").json()
response
```

```python
response = tiny.market_data.get_company_dividends(symbols="MSFT,MO,KMI")
response
```

```python
import pandas as pd
df_div = pd.DataFrame()
response.json()
for r in response.json():
    for result in r['results']:
        df_temp = pd.DataFrame(result['tables']['cash_dividends'])
        df_temp['symbol'] = r['request']
        df_temp['type'] = result['type']
        df_div = pd.concat([df_div,df_temp],ignore_index=True)
df_div.shape
```
